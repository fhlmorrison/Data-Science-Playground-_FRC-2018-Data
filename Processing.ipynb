```python
import pandas as pd
```

```python
data = pd.read_csv("TestDataFRC2018.csv", header=2, sep=" *,", engine="python")
```

```python
# Remove blank rows
print(f"Number of rows before cleaning: {data.shape[0]}")
data.dropna(axis=0,inplace=True)
f"Number of rows after cleaning: {data.shape[0]}"
```

```python
data['Match Number'] = data['Match Number'].astype(int)
data['Team Number'] = data['Team Number'].astype(int)
data.info()
```

```python
renameDict = {
    "Switch" : "Auto Switch",
    "Scale" : "Auto Scale",
    "Exchange" : "Auto Exchange",
    "Scale.1" : "Teleop Scale",
    "Exchange.1" : "Teleop Exchange",
}
data.rename(columns=renameDict, inplace=True)
```

```python
columnsToDrop = [
    "Date Time",
    "Time Stamp",
    "Alliance Colour",
    "Match Number"
]
data.drop(columns=columnsToDrop, inplace=True)
```
